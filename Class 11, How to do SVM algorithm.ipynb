```python
import pandas as pd

df = pd.read_csv('Heart Disease Dataset (1).csv')
df
```

```python
df.info()
```

```python
import seaborn as sns
sns.FacetGrid(df, hue = 'target', height = 5).map(sns.kdeplot, 'chol').add_legend()
```

```python
import seaborn as sns
sns.FacetGrid(df, hue = 'target', height = 5).map(sns.kdeplot, 'cp').add_legend()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (12,10))

corr = df.corr()
sns.heatmap(corr, annot = True, cmap = plt.cm.CMRmap_r)
plt.show()
```

```python
feature_cols = ['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal','target']

x = df[feature_cols]
y = df['target']


```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.10)

x_test
```

```python
x_test.info()
```

```python
from sklearn.svm import SVC

svm = SVC(kernel = 'linear')

svm.fit(x_train, y_train)

Y_pred = svm.predict(x_test)
Y_pred
```

```python
y_test
```

```python
from sklearn.svm import SVC

svm = SVC(kernel = 'rbf')

svm.fit(x_train, y_train)

Y_pred = svm.predict(x_test)
Y_pred
```

```python
y_test
```

```python
from sklearn import model_selection

kfold = model_selection.KFold(n_splits = 10, random_state = 31)
model = SVC(kernel = 'linear')
results = model_selection.cross_val_score(model, x_train, y_train, cv =kfold) 
results
```

```python
from sklearn import metrics
print('Accuracy: ', metrics.accuracy_score(y_test, Y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, Y_pred))
```

```python
from sklearn import tree

tree = tree.DecisionTreeClassifier()
tree.fit(x_train, y_train)
Y_pred = tree.predict(x_test)
Y_pred
```

```python
metrics.accuracy_score(y_test, Y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 100)
rf.fit(x_train, y_train)
Y_pred = rf.predict(x_test)
Y_pred
```
