```python
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt
import pandas as pd

data = [10,12,9,12,6,5,16,12,15,11,13,15,5,18,11,17,12,14,8,8,21,13,5,13,15,11,12,23]

df = pd.DataFrame(data)

nresults = seasonal_decompose(df, model = 'multitative', freq = 4)
nresults.plot()
plt.show()
```

```python
plt.plot(list(range(1,29)),pd.Series(nresults.trend[0]))
```

```python
nresults.trend[0]
```

```python
plt.plot(list(range(1,29)),pd.Series(nresults.seasonal[0]))
```

```python
plt.plot(list(range(1,29)),pd.Series(nresults.resid[0]))
```

```python
plt.bar(list(range(1,29)),pd.Series(nresults.resid[0]))
```
