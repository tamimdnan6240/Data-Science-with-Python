```python
#As there is no outlier in our code, so do again after putting an outlier

import pandas as pd

Weight_Data = [['tom', 0.3],['pumpkin',3.5],['bonk',4.2],['thunder',5.50],['oreo',4.73],['nya',5.0],['kitkat',4.55],['bubbles',4.90],['sparkle',6.29],['pebbles',3.72]]

df_weight = pd.DataFrame(Weight_Data, columns = ['Name', 'Weight(Kg)'])
df_weight

sns.boxplot(y = "Weight(Kg)", data = df_weight)
```

```python
#now find the value of the outliers by calculating percentiles

Q1 = np.percentile(df_weight['Weight(Kg)'], 25)
Q3 = np.percentile(df_weight['Weight(Kg)'], 25)

IQR = Q3 - Q1

low = Q1 - 1.5 * IQR
up = Q3 + 1.5 * IQR

#make a list named  to store values obtained from the loop

outlier = []

for x in df_weight['Weight(Kg)'] :
  if ((x>up) or (x<low)):
    outlier.append(x)

print("The outliers are",outlier)
```
